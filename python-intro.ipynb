<h2>0. Introdução</h2>

<h3>a. Ciência da Computação</h3>

<p> A Ciência da Computação costuma ser muito associada a computadores, apesar de não ser o estudo deles propriamente ditos. Ela é, no geral, o estudo de problemas e da existência - ou não - de soluções para eles. É conhecimento fundamental pra alguém que estuda Ciência da Computação saber comparar essas soluções, dividir em tarefas menores e até já saber alguns resutados que facilitam o caminho.

<p> Um cientista da computação é alguém que, tendo um problema, deve desenvolver um raciocínio para achar uma resposta. Esse desenvolvimento é organizado numa sequência de passos, de instruções, que, se seguidas, irão resolver o problema, e essa sequência recebe o nome algoritmo. Mais formalmente, um algoritmo é uma sequência finita e bem definida de passos computacionais que recebe uma entrada (um input) e retorna uma saída (um output).

<p> Um mesmo exercício pode ser resolvido de maneiras diferentes, algumas melhores e outras piores. Mas como avaliar se uma solução é boa ou ruim? Um critério possível é: chegar no mesmo resultado mais rapidamente e usando menos recursos indica uma boa solução. Dois algoritmos podem estar certos, mas um pode ser melhor que o outro.

<p> Algo que surge quando falamos em ciência da computação é a palavra computável. Um problema é computável se existe um algoritmo para resolvê-lo. Existem, inclusive, problemas que não podem ser resolvidos seguindo um algoritmo, problemas não computáveis.

<p> Este material usará vários conceitos de ciência da computação, mas não necessariamente irá definir cada um deles. O importante aqui é aprender o básico de como pensar em um algoritmo, como desenvolver uma lógica para resolver um problema.

<h3>b. Programação</h3>

<p> Programação é, basicamente, uma forma de transcrever um algoritmo para uma linguagem de programação, fazendo com que essa sequência possa ser executada por um computador. A programação é simplesmente o meio, uma ferramenta que nós usamos para traduzir nossa lógica para solucionar o problema.

<p> Ciência da computação, como já dito, não é a ciência de programar, mas programar vai ser algo muito importante no processo. É programando que nós aplicamos nossos algoritmos, e nós vamos inclusive falar de várias técnicas de programação nesse material, mas note que o algoritmo vem primeiro, só depois vem a programação. Sem um algoritmo para ser transcrito, um programa não existe.


<h3>c. O que é um programa</h3>

<p>Um programa é uma sequência de instruções que um computador pode executar para completar uma tarefa específica. Atualmente, escrevemos nossos programas utilizando uma linguagem de programação, mas programas já foram escritos até mesmo usando cartões perfurados.</p>


<i>Cartão perfurado usado pra carregar software em um computador antigo. <a href="https://www.flickr.com/photos/binaryape/5151286161/" /></i> Fonte

<p>Os computadores modernos funcionam por meio de <i>transistores</i>, que são como interruptores que deixam ou não passar corrente elétrica por eles; seu celular tem bilhões desses transistores. A linguagem que o computador entende "naturalmente" é binária: 1 ou 0, corrente passando ou corrente não passando.</p>

<p>Antigamente, para dar instruções aos computadores, os <i>programadores</i> (pessoa que programa um computador) deviam escrever o código todo em <i>código de máquina</i>, com as instruções diretamente na forma que o <i>processador</i> (parte do computador responsável por fazer cálculos) foi projetado para entender. Um programa muito simples em código de máquina se parece com isto:</p>
<br/ >
<code>B4 03 CD 10 B0 01 B3 0A B9 0B 00 BD 13 01 B4 13<br>
CD 10 C3 4F 69 20 6D 75 6E 64 6F 21 0D 0A</code>
<br />
<p>Não é difícil convencer alguém que essa não é uma forma muito prática (e principalmente muito intuitiva) de se programar. Para resolver esse problema é que foram criadas as <i>linguagens de programação</i>.</p>

<h3>d. O que é uma linguagem</h3>

<p><i>Linguagens naturais</i> são aquelas que surgiram e se desenvolveram com o passar do tempo, como o Inglês, o Português ou o Francês. Não conseguimos pensar em um momento específico em que esse tipo de linguagem foi inventada, ela foi sendo criada e moldada conforme as pessoas a falavam, além de estarem em constante mudança.</p>
<p><i>Linguagens formais</i> são criadas para aplicações específicas, como a linguagem matemática para representar relações entre números e outros objetos matemáticos. Esse tipo de linguagem costuma ter uma estrutura estrita, com regras de sintaxe que definem se uma declaração é válida ou não.</p>
<p><i>Linguagens de programação</i> são linguagens formais feitas especificamente para dar instruções a um computador. Apesar disso, também estão sempre em mudança e de tempos em tempos surgem novas linguagens de programação de acordo com as necessidades que surgem. </p>

<h3>a. Olá, mundo!</h3>

```python
print('Olá, mundo!')
```

<p> Esse costuma ser o primeiro programa de todos que se aventuram pelo mundo da programação de computadores. Vamos quebrar esse código e tentar entendê-lo.
<p> O termo print() é o que chamamos de <i>função</i>, um conceito bem semelhante àquele que vemos na matemática. Podemos imaginar uma função como sendo uma máquina em que colocamos algo dentro (chamaremos de <i>argumento</i> ou <i>input</i>), ela transforma isso em outra coisa e nos devolve o produto da transformação (<i>saída</i> ou <i>output</i>). Em pyhton, as funções sempre serão algo parecido com nome_da_função(argumento).
<p> No caso de print('Olá, mundo!'), o nome é justamente print, o argumento é 'Ólá, mundo!' e a saída é mostrar (imprimir) na tela o que está em parênteses.
<p> Uma ferramenta muito útil é a função help(). Ela recebe como argumento outra função do python e retorna parte da <i>documentação</i> daquela que está no argumento. Em outras palavras, help() te ajuda dizendo como a função dentro dos parênteses funciona, quais parâmetros ela recebe e o que ela gera como output. Vamos ver como se comporta print().

```python
help(print)
```

<p>  Note que ela não tem apenas um argumento, mas sim uma lista deles! O que acontece é que, em pyhton, alguns desses argumentos são opcionais e, se não atribuirmos nenhum valor a eles, o próprio pyhton confere um valor padrão.
<p> Poder preencher apenas os valores que nos interessam economiza bastante tempo, e existem situações onde queremos (ou precisamos) especificar outros também. Por exemplo, através do argumento <i>end</i>, podemos definir o que a função vai imprimir no fim dela.
<p> Por padrão, end recebe '\n', descrito como "string appended after the last value, default a newline". Em outras palavras, a função print() pula para a próxima linha quando o que foi impresso chega ao final. De fato:

```python
print('Olá')
print('Olá')
```

<p> Vamos, agora, mudar o valor do argumento end e ver como o output será afetado:

```python
print('Olá, mundo!', end= '---')
```

```python
print('Olá', end= ' ')
print('Olá', end= ' ')
```

<p> Apesar de aparecer, implicitamente e por padrão, no final da mensagem impressa na tela, o termo '\n' significa, basicamente, "pule uma linha" e podemos utilizá-lo em outras partes da função:

```python
print('O\nlá')
```

```python
print('Ol\ná')
```

<p> Pelo que parece, '\n' irá pular uma linha sempre que estiver entre aspas. Mas, atenção: em python, 'olá' e "olá" são equivalentes, é sua escolha qual usar. Entretanto, não é permitido usar 'olá"! Sempre que começar com um, deve-se terminar com o mesmo.
<p> Com a print(), podemos imprimir mais de um valor na tela. Procedemos da seguinte forma: print(valor1, valor2). Dessa forma, a função irá imprimir os dois valores separados por um espaço:

```python
print('Cachorros', 'são legais')
```

<h3>b. Variáveis e Tipos</h3>

<h>

<p> Nesta seção iremos introduzir as noções de <i>variáveis</i> (basicamente, algo que mantém um certo valor enquanto você queira) e de <i>tipos de dados</i> (como um dado irá se comportar).

<h>

<h4>Tipos</h4>

<p> Os dados que usamos em python possuem tipos e cada um irá se comportar de acordo com o tipo em que ele se enquadrar. Uma forma fácil de descobrir o tipo de um dado é utilizando a função type(nome_do_dado). Por exemplo:

```python
type(5)
```

```python
type('5')
```

<p> Vemos, então, que o dado 5 é do tipo int, enquanto que o dado '5' é do tipo str. Posteriormente, veremos o que significam esses nomes.
<p> Aqui, caso você esteja utilizando o caderno do Jupyter (ou outro parecido), cabe uma observação: quando uma célula é executada e vemos "Out[x]:", é importante lembrar que ele <b>não está sendo mostrado para o usuário, impresso na tela</b>. "Out[x]:" serve apenas para o programador ver o resultado do que foi feito naquela célula. Caso quem fez o programa queira, de fato, mostrar algo ao usuário, é preciso utilizar a função print().
<p> Voltando para os tipos, os principais são:

    1. Inteiros (int): são os números inteiros, como 1, 2, -3 e 0 (sem a parte decimal)
    
    2. Números de ponto flutuante (float): são os números que têm parte decimal, como 1.3, 2.7, -1.5 e 0.2;
    
    3. Strings (str): são basicamente uma sequência de caracteres, sendo representados, em python, por qualquer coisa entre aspas;
    
    4. Booleanos (bool): são aqueles dados que podem assumir exatamente dois valores, como verdadeiro (True) e falso (False).

<p> Vejamos, agora, exemplos desses 4 tipos de variáveis, na sequência mostrada:

```python
print(1 + 2)
```
