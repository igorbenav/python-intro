<h3>a. Ciência da Computação</h3>

<p>Ciência da Computação normalmente é algo muito associado com computadores, mas não é o estudo de computadores. Ciência da Computação é, no geral, o estudo de problemas, nisso temos, além deles, as soluções - ou não existência de soluções - pra esses problemas. É conhecimento fundamental pra alguém que estuda Ciência da Computação saber comparar essas soluções, dividir em tarefas menores e até já saber algumas soluções que facilitam o caminho.</p>

<p>Um cientista da computação é alguém que, tendo um problema, desenvolve uma lógica para achar uma solução. Essa lógica vem numa sequência de passos, de instruções, que se seguidas vão resolver esse problema, e é essa sequência que nós chamamos de <b>algoritmo</b>. Um algoritmo é uma sequência finita bem definida de passos computacionais que recebe uma entrada (um <i><b>input</b></i>) e retorna uma saída (um <i><b>output</b></i>).</p>

<p>Um mesmo problema pode ser resolvido de maneiras diferentes, algumas maneiras melhores e outras piores. Mas o que seria uma maneira melhor de resolver um problema? Se você chega no mesmo resultado mais rapidamente e usando menos recursos você tem uma maneira melhor. Dois algoritmos podem estar certos, mas um pode ser melhor que o outro.</p>

<p>Uma palavra que surge quando falamos de ciência da computação é a palavra <b>computável</b>. Um problema é computável se existe um algoritmo para resolvê-lo. Existem, inclusive, problemas que não podem ser resolvidos seguindo um algoritmo, problemas não computáveis.</p>

<p>Esse material vai usar vários conceitos de ciência da computação, mas não necessariamente vai definir cada coisa. O importante aqui é aprender o básico de como pensar em um algoritmo, como desenvolver uma lógica para resolver um problema.</p>

<h3>b. Programação</h3>

<p>Programação é basicamente uma forma de transcrever um algoritmo para uma linguagem de programação, fazendo que essa sequência possa ser executada por um computador. A programação é simplesmente o meio, uma ferramenta que nós usamos para traduzir nossa lógica para resolver o problema.</p>

<p>Ciência da computação, como já dito, não é a ciência de programar, mas programar vai ser algo muito importante no processo. É programando que nós aplicamos nossos algoritmos, e nós vamos inclusive falar de várias técnicas de programação nesse material, mas note que o algoritmo vem primeiro, só depois vem a programação. Sem um algoritmo para ser transcrito, um programa não existe.</p>

<h3>c. O que é um programa</h3>

<p>Um programa é uma sequência de instruções que um computador pode executar para completar uma tarefa específica. Normalmente escrevemos um programa usando uma linguagem de programação, mas programas já foram escritos até mesmo usando cartões perfurados.</p>


<i>Cartão perfurado usado pra carregar software em um computador antigo. <a href="https://www.flickr.com/photos/binaryape/5151286161/" /></i> Fonte

<p>Os computadores modernos funcionam por meio de <i>transistores</i>, que são como interruptores que deixam ou não passar corrente elétrica por eles, seu celular tem bilhões desses transistores. A linguagem que o computador entende "naturalmente" é binário: 1 ou 0, corrente passando ou corrente não passando.</p>

<p>Para dar instruções para computadores antigamente, os <i>programadores</i> (pessoa que programa um computador) deviam escrever o código todo em <i>código de máquina</i>, com as instruções diretamente da forma que o <i>processador</i> (parte do computador responsável por fazer cálculos) foi projetado para entender. Um programa muito simples em código de máquina se parece com isto:</p>
<br/ >
<code>B4 03 CD 10 B0 01 B3 0A B9 0B 00 BD 13 01 B4 13<br>
CD 10 C3 4F 69 20 6D 75 6E 64 6F 21 0D 0A</code>
<br />
<p>Não é difícil convencer que essa não é uma forma muito prática (e principalmente muito intuitiva) de se programar. Para resolver esse problema que foram criadas as <i>linguagens de programação</i>.</p>

<h3>d. O que é uma linguagem</h3>

<p><i>Linguagens naturais</i> são as linguagens que surgiram naturalmente, como Inglês, Português, Francês. Não conseguimos pensar em um momento que esse tipo de linguagem foi inventada, ela foi evoluindo e surgindo com as pessoas que a falam.</p>
<p><i>Linguagens formais</i> são feitas por pessoas para aplicações específicas, como a linguagem matemática para representar relações entre números e outros objetos matemáticos. Esse tipo de linguagem costuma ter uma estrutura estrita, com regras de sintaxe que definem se uma declaração é válida ou não.</p>
<p><i>Linguagens de programação</i> são linguagens formais feitas especificamente para dar instruções a um computador.</p>
